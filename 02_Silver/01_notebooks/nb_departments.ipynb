```python
from pyspark.sql.functions import trim, col, monotonically_increasing_id

departments_df = spark.read.table("global_tech_bronze.global_tech_schema.raw_departments")

departments_df = (
    departments_df.withColumn("department_code", trim(col("department_code")))
    .withColumn("cost_center", trim(col("cost_center")))
    .withColumn("department_name", trim(col("department_name")))
    .withColumn("department_sk",monotonically_increasing_id())
)

departments_df.write.format("delta").mode("overwrite").saveAsTable("global_tech_sliver.transformed_data_silver.tf_departments")

display(departments_df)
```

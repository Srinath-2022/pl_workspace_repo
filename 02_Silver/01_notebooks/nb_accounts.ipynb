```python
from pyspark.sql.functions import trim,monotonically_increasing_id, col

accounts_df = spark.read.table("global_tech_bronze.global_tech_schema.raw_accounts")
accounts_df= accounts_df.withColumn("account_name", trim("account_name")).withColumn("account_code", trim("account_code")).withColumn("account_id", monotonically_increasing_id()).withColumn("category", trim("category")).withColumnRenamed("is_active", "account_is_active")

accounts_df.write.format("delta").mode("overwrite").saveAsTable("global_tech_sliver.transformed_data_silver.tf_accounts")
   
display(accounts_df)

```

```python
%sql
CREATE OR REPLACE TABLE global_tech_gold.global_tech_cube_table.unified_kpi_cube AS
SELECT
  'FINANCE' AS record_type,
  f.company_id,
  f.department_id,
  CAST(NULL AS BIGINT) AS employee_id,
  f.posting_date,
  f.amount,
  f.account_type,
  f.category,
  CAST(NULL AS DOUBLE) AS total_compensation,
  CAST(NULL AS DOUBLE) AS bonus,
  CAST(NULL AS DOUBLE) AS overtime_pay,
  e.position,
  e.termination_date,
  d.department_name,
  c.company_name
FROM global_tech_gold.global_tech_fact_tables.fact_finance f
LEFT JOIN global_tech_gold.global_tech_dim_tables.dim_employee e ON f.company_id = e.employee_id
LEFT JOIN global_tech_gold.global_tech_dim_tables.dim_department d ON f.department_id = d.department_id
LEFT JOIN global_tech_gold.global_tech_dim_tables.dim_company c ON f.company_id = c.company_id

UNION ALL

SELECT
  'PAYROLL' AS record_type,
  p.company_id,
  p.department_id,
  p.employee_id,
  p.pay_date AS posting_date,
  CAST(NULL AS DOUBLE) AS amount,
  CAST(NULL AS STRING) AS account_type,
  CAST(NULL AS STRING) AS category,
  p.total_compensation,
  p.bonus,
  p.overtime_pay,
  e.position,
  e.termination_date,
  d.department_name,
  c.company_name
FROM global_tech_gold.global_tech_fact_tables.fact_payroll p
LEFT JOIN global_tech_gold.global_tech_dim_tables.dim_employee e ON p.employee_id = e.employee_id
LEFT JOIN global_tech_gold.global_tech_dim_tables.dim_department d ON p.department_id = d.department_id
LEFT JOIN global_tech_gold.global_tech_dim_tables.dim_company c ON p.company_id = c.company_id;

SELECT * FROM global_tech_gold.global_tech_cube_table.unified_kpi_cube;
```
